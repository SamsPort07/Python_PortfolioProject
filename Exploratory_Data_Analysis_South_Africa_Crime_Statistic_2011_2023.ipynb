# Analysis and Inference Thus far:
We can confidently say our data from the sheet 'Crime incidents by category' is clean, well documented, and well structured.

We can summarize the data as follows:
1. The Data categorises Geography's, associated crime category, for a given financial year, if the Geography is ZA the Count  is the cummulative summarize off all other Geography's for the given Financial Year. 'Count' is the number of occurences of a specific crime category for a given year.

Let's begin plotting our data in an effort to gain deeper insight and understanding.

```python
CrimeIncidents_df['Crime Category'].unique()
```

# Visualization

```python
# The X-axis will be the financial years.
# The Y-axis will be the counts of a given crime category.

years = CrimeIncidents_df[(CrimeIncidents_df.Geography == "ZA")]['Financial Year'].unique()

contactCrimeCount = CrimeIncidents_df[(CrimeIncidents_df.Geography == "ZA") & (CrimeIncidents_df['Crime Category'] == "Contact Crimes")].Count.values
sexualCrimeCount = CrimeIncidents_df[(CrimeIncidents_df.Geography == "ZA") & (CrimeIncidents_df['Crime Category'] == "Sexual Offences")].Count.values
policeActionCrimeCount = CrimeIncidents_df[(CrimeIncidents_df.Geography == "ZA") & (CrimeIncidents_df['Crime Category'] == "Crimes Detected as a Result of Police Action")].Count.values
aggravatedRobberiesCount = CrimeIncidents_df[(CrimeIncidents_df.Geography == "ZA") & (CrimeIncidents_df['Crime Category'] == "Aggravated Robberies")].Count.values
contactRelatedCrimeCount = CrimeIncidents_df[(CrimeIncidents_df.Geography == "ZA") & (CrimeIncidents_df['Crime Category']) == 'Contact Related Crimes'].Count.values

plt.figure(figsize=(15,8))
plt.xticks(rotation=90)

plt.plot(years, contactCrimeCount,'o-b')
plt.plot(years, sexualCrimeCount, 's--r')
plt.plot(years, policeActionCrimeCount, '*:c')
plt.plot(years, aggravatedRobberiesCount, 'h-k')
#plt.plot(years, contactRelatedCrimeCount,'p--y')


plt.xlabel('Financial Year')
plt.ylabel('Count of Crimes for Various Categories')
plt.title('Count of Crimes of Various Nature in ZA over Years')
plt.legend(["CC", "SO", "PAC", "ARC", "CRC"],loc="upper right",borderaxespad=0)

plt.show();
```
